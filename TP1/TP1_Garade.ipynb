Para las imágenes que serán proporcionadas de Tito se pide:

1. Implementar el algoritmo de pasaje a coordenadas cromáticas para librarnos de las variaciones de contraste.
2. Implementar el algoritmo White Patch para librarnos de las diferencias de color de iluminación 

```python
# Importar librerías
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

## Ejercicio 1

```python
def img_to_chrom_coord(img):
    img1_new = np.zeros(shape=img.shape)
    img1_new[:,:,0] = img[:,:,0] / np.sum(img, axis=2)
    img1_new[:,:,1] = img[:,:,1] / np.sum(img, axis=2)
    img1_new[:,:,2] = img[:,:,2] / np.sum(img, axis=2)
    return img1_new
```

```python
img1 = cv.imread('CoordCrom_1.png')
print('\nTamaño imagen: {} pix'.format(img1.shape))
img1RGB = cv.cvtColor(img1,cv.COLOR_BGR2RGB)
img1_new = img_to_chrom_coord(img1RGB)
plt.subplots(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(img1RGB)
plt.subplot(1,2,2)
plt.imshow(img1_new)
plt.show()
```

```python
img2 = cv.imread('CoordCrom_2.png')
print('\nTamaño imagen: {} pix'.format(img2.shape))
img2RGB = cv.cvtColor(img2,cv.COLOR_BGR2RGB)
img2_new = img_to_chrom_coord(img2RGB)
plt.subplots(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(img2RGB)
plt.subplot(1,2,2)
plt.imshow(img2_new)
plt.show()
```

## Ejercicio 2

```python
def white_patch(img):
    max_red = np.max(img[:,:,0])
    max_green = np.max(img[:,:,1])
    max_blue = np.max(img[:,:,2])
    img_new = np.zeros(shape=img.shape)
    img_new[:,:,0] = img[:,:,0] / max_red
    img_new[:,:,1] = img[:,:,1] / max_green
    img_new[:,:,2] = img[:,:,2] / max_blue
    return img_new
```

```python
test_blue = cv.imread('test_blue.png')
print('\nTamaño imagen: {} pix'.format(test_blue.shape))
test_blueRGB = cv.cvtColor(test_blue,cv.COLOR_BGR2RGB)
test_blue_new = white_patch(test_blueRGB)
plt.subplots(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(test_blueRGB)
plt.subplot(1,2,2)
plt.imshow(test_blue_new)
plt.show()
```
